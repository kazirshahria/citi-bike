# Haversine Formula

[Credit](https://www.geeksforgeeks.org/dsa/haversine-formula-to-find-distance-between-two-points-on-a-sphere/)

```python
import pandas as pd
import math
```

```python
df = pd.read_csv('../data/2025-citibike-clean.csv', index_col=[0])
df.head()
```

```python
def haversine(lat1, lon1, lat2, lon2):
    dLat = (lat2 - lat1) * math.pi / 180.0
    dLon = (lon2 - lon1) * math.pi / 180.0

    lat1 = (lat1) * math.pi / 180.0
    lat2 = (lat2) * math.pi / 180.0

    a = (pow(math.sin(dLat / 2), 2) + 
         pow(math.sin(dLon / 2), 2) * 
             math.cos(lat1) * math.cos(lat2));
    rad = 6371
    c = 2 * math.asin(math.sqrt(a))
    return rad * c * 0.6213711922
```

```python
df['miles'] = df.apply(lambda col: haversine(col['start_lat'], col['start_lng'], col['end_lat'], col['end_lng']), axis=1)
```

```python
df.to_parquet('../data/2025-citibike.parquet')
```
